```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Helper Functions
def add_time_features(df):
    df['hour'] = df['timestamp'].dt.hour
    df['dayofweek'] = df['timestamp'].dt.dayofweek
    df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
    df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
    df['day_sin'] = np.sin(2 * np.pi * df['dayofweek'] / 7)
    df['day_cos'] = np.cos(2 * np.pi * df['dayofweek'] / 7)
    return df

def create_lag_features(df, target_col='equipment_energy_consumption', lags=[1, 2, 3]):
    df = df.sort_values('timestamp')
    for lag in lags:
        df[f'{target_col}_lag{lag}'] = df[target_col].shift(lag)
    return df

def create_interactions(df):
    if 'zone1_temperature' in df.columns and 'zone1_humidity' in df.columns:
        df['temp_humidity_interaction'] = df['zone1_temperature'] * df['zone1_humidity']
    return df

def create_aggregates(df):
    temp_cols = [col for col in df.columns if 'temperature' in col]
    humid_cols = [col for col in df.columns if 'humidity' in col]
    df['avg_zone_temp'] = df[temp_cols].mean(axis=1)
    df['avg_zone_humidity'] = df[humid_cols].mean(axis=1)
    return df

def log_transform_target(y): return np.log1p(y)
def inverse_log_transform(y_log): return np.expm1(y_log)

def calculate_vif(X, thresh=5.0):
    X_vif = X.copy()
    dropped = True
    while dropped:
        dropped = False
        vif_data = pd.DataFrame()
        vif_data["feature"] = X_vif.columns
        vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]
        max_vif = vif_data['VIF'].max()
        if max_vif > thresh:
            drop_feature = vif_data.loc[vif_data['VIF'].idxmax(), 'feature']
            X_vif.drop(columns=[drop_feature], inplace=True)
            dropped = True
    return X_vif

# Load and Preprocess Data
df = pd.read_csv("/content/data (1).csv")
df['timestamp'] = pd.to_datetime(df['timestamp'])
for col in df.columns:
    if col != 'timestamp':
        df[col] = pd.to_numeric(df[col], errors='coerce')
df = df[(df['lighting_energy'] >= 0) & (df['equipment_energy_consumption'] >= 0)]
for col in df.columns:
    if col != 'timestamp':
        df[col].fillna(df[col].median(), inplace=True)

# Feature Engineering
df = add_time_features(df)
df = create_lag_features(df)
df = create_interactions(df)
df = create_aggregates(df)
df.dropna(inplace=True)

# Correlation Heatmap
plt.figure(figsize=(12, 10))
corr_matrix = df.drop(columns='timestamp').corr()
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()

# Define X and y
X = df.drop(columns=['timestamp', 'equipment_energy_consumption'])
y = df['equipment_energy_consumption']

# Multicollinearity Check
X = calculate_vif(X)

# Anomaly Detection using IQR
Q1, Q3 = y.quantile([0.25, 0.75])
IQR = Q3 - Q1
lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
mask = (y >= lower) & (y <= upper)
y_clean = y.copy()
y_clean[~mask] = y.median()
X_clean = X.copy()
X_clean[~mask] = np.nan
X_clean.fillna(X_clean.median(), inplace=True)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42)
y_train_log = log_transform_target(y_train)
y_test_log = log_transform_target(y_test)

# Base Models
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(random_state=42),
    "XGBoost": XGBRegressor(random_state=42)
}

# Cross-validation, Training and Evaluation
for name, model in models.items():
    print(f"\n {name} Results:")
    scores = cross_val_score(model, X_train, y_train_log, cv=5, scoring='r2')
    print("Cross-validated R²:", np.round(scores, 4))
    model.fit(X_train, y_train_log)

    # Train predictions
    y_train_pred_log = model.predict(X_train)
    y_train_pred = inverse_log_transform(y_train_pred_log)
    y_train_true = inverse_log_transform(y_train_log)
    print("Train R²:", round(r2_score(y_train_true, y_train_pred), 4))
    print("Train MAE:", round(mean_absolute_error(y_train_true, y_train_pred), 2))

    # Test predictions
    y_test_pred_log = model.predict(X_test)
    y_test_pred = inverse_log_transform(y_test_pred_log)
    y_test_true = inverse_log_transform(y_test_log)
    print("Test R²:", round(r2_score(y_test_true, y_test_pred), 4))
    print("Test MAE:", round(mean_absolute_error(y_test_true, y_test_pred), 2))

# Hyperparameter Tuning - XGBoost
xgb_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.05, 0.1]
}
xgb_gs = GridSearchCV(XGBRegressor(random_state=42), xgb_param_grid, cv=3, scoring='r2')
xgb_gs.fit(X_train, y_train_log)
print("\n Best XGBoost Params:", xgb_gs.best_params_)

best_xgb = xgb_gs.best_estimator_
y_pred_final_xgb = inverse_log_transform(best_xgb.predict(X_test))
print("\n Final XGBoost Evaluation")
print("R² Score:", round(r2_score(y_test_true, y_pred_final_xgb), 4))
print("MAE     :", round(mean_absolute_error(y_test_true, y_pred_final_xgb), 2))

# Hyperparameter Tuning - Random Forest
rf_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
rf_gs = RandomizedSearchCV(RandomForestRegressor(random_state=42), rf_param_grid, n_iter=10, cv=3, scoring='r2', random_state=42)
rf_gs.fit(X_train, y_train_log)
print("\n Best Random Forest Params:", rf_gs.best_params_)

best_rf = rf_gs.best_estimator_
y_pred_final_rf = inverse_log_transform(best_rf.predict(X_test))
print("\n Final Random Forest Evaluation")
print("R² Score:", round(r2_score(y_test_true, y_pred_final_rf), 4))
print("MAE     :", round(mean_absolute_error(y_test_true, y_pred_final_rf), 2))

# Plot Predicted vs Actual
plt.figure(figsize=(6, 6))
plt.scatter(y_test_true, y_pred_final_xgb, alpha=0.5, label='XGBoost')
plt.scatter(y_test_true, y_pred_final_rf, alpha=0.5, label='Random Forest')
plt.plot([y_test_true.min(), y_test_true.max()], [y_test_true.min(), y_test_true.max()], 'k--')
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Predicted vs Actual (Tuned Models)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
